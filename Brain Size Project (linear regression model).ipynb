```python
import statsmodels.formula.api as smf
```

```python
model = smf.ols("PIQ ~ Brain + Height + Weight", data=brain_data).fit()
```

```python
print(model.summary())
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
model = smf.ols('PIQ  ~ Brain + Height + Weight +Brain*Height', data=brain_data).fit()
anova_table = sm.stats.anova_lm(model, type=2)
```

```python
print(anova_table)
```
